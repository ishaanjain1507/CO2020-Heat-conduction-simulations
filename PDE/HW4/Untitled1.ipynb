```python
import numpy as np
import matplotlib.pyplot as plt
from math import log
```

```python
err = np.array([2.01437e-04, 5.10404e-05, 1.28431e-05, 3.18317e-06])
err2 = np.array([log(i) for i in err])
# x = np.array([log(i) for i in range(1,5)])
x = np.arange(1,5,1)
x
```

```python
fig = plt.figure()
plt.plot(x, err2, 'o-')
plt.xticks(x,["17*9", "33*17", "65*33", "129*65"])
plt.yticks(err2, [f"log({i})" for i in err])
plt.title("variation of error with grid size")
plt.grid(True)
plt.xlabel("grid size")
plt.ylabel("log of error")
```
