```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Tex = np.array([])
```

```python
import pandas as pd
```

```python
nx = 33
ny = 17
df = pd.read_csv("./T_xy_033_017_0000.csv")
df
```

```python
num_arr = df.to_numpy()
```

```python
T = num_arr[:,2].reshape(nx,ny).T
print(T)
Tex = num_arr[:,3].reshape(nx,ny).T
print(Tex)
```

```python

y = np.linspace(0, 0.5, ny)
x = np.linspace(0,1,nx)
x,y
```

```python
plt.contourf(x,y,T, cmap = "inferno")
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Variation of T in a {T.shape[1]}*{T.shape[0]} grid")
plt.colorbar()
plt.savefig(f"{nx}*{ny}.png")
```

```python
plt.contourf(x,y,Tex, cmap = "inferno")
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Variation of T(exact solution) in a {Tex.shape[1]}*{Tex.shape[0]} grid")
plt.colorbar()
plt.savefig(f"{nx}*{ny} exact.png")
```
